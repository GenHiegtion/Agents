### __Generator__

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.messages import AIMessage, HumanMessage
from langchain_openai import ChatOpenAI

generation_prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "Báº¡n lÃ  trá»£ lÃ½ cho má»™t Tech Influencer (ngÆ°á»i cÃ³ táº§m áº£nh hÆ°á»Ÿng vá» cÃ´ng nghá»‡) trÃªn Twitter, nhiá»‡m vá»¥ cá»§a báº¡n lÃ  viáº¿t nhá»¯ng bÃ i tweet xuáº¥t sáº¯c."
            " HÃ£y táº¡o ra bÃ i tweet hay nháº¥t cÃ³ thá»ƒ dá»±a trÃªn yÃªu cáº§u cá»§a ngÆ°á»i dÃ¹ng."
            " Náº¿u ngÆ°á»i dÃ¹ng Ä‘Æ°a ra lá»i phÃª bÃ¬nh, hÃ£y pháº£n há»“i báº±ng má»™t phiÃªn báº£n Ä‘Ã£ Ä‘Æ°á»£c chá»‰nh sá»­a láº¡i tá»« nhá»¯ng láº§n thá»­ trÆ°á»›c Ä‘Ã³.",
        ),
        MessagesPlaceholder(variable_name="messages"),
    ]
)

llm = ChatOpenAI(model="gpt-4o-mini")
generate_chain = generation_prompt | llm
```

```python
request = HumanMessage(
    content="HÃ£y viáº¿t má»™t bÃ i Ä‘Äƒng trÃªn Twitter vá» lÃ½ do táº¡i sao HoÃ ng tá»­ bÃ© váº«n cÃ²n phÃ¹ há»£p vá»›i tuá»•i thÆ¡ hiá»‡n Ä‘áº¡i."
)
response = generate_chain.invoke({"messages": [request]})
response.content
```

### __Reflector__

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_openai import ChatOpenAI

reflection_prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "Báº¡n lÃ  má»™t Twitter Influencer Ä‘Ã¬nh Ä‘Ã¡m Ä‘ang cháº¥m Ä‘iá»ƒm má»™t bÃ i tweet. HÃ£y Ä‘Æ°a ra lá»i phÃª bÃ¬nh vÃ  cÃ¡c Ä‘á» xuáº¥t cáº£i thiá»‡n cho bÃ i tweet cá»§a ngÆ°á»i dÃ¹ng."
            " LuÃ´n cung cáº¥p cÃ¡c Ä‘á» xuáº¥t chi tiáº¿t, bao gá»“m yÃªu cáº§u vá» Ä‘á»™ dÃ i, kháº£ nÄƒng viral, vÄƒn phong, v.v.",
        ),
        MessagesPlaceholder(variable_name="messages"),
    ]
)

reflect_chain = reflection_prompt | llm
```

```python
reflected_response = reflect_chain.invoke({"messages": [request, HumanMessage(content=response.content)]})
reflected_response.content
```

### __Graph__

```python
from typing import List, Sequence
from langchain_core.messages import BaseMessage
from langgraph.graph import END, MessageGraph

REFLECT = "reflect"
GENERATE = "generate"


def generation_node(state: Sequence[BaseMessage]):
    return generate_chain.invoke({"messages": state})


def reflection_node(messages: Sequence[BaseMessage]) -> List[BaseMessage]:
    res = reflect_chain.invoke({"messages": messages})
    return [HumanMessage(content=res.content)]


builder = MessageGraph()
builder.add_node(GENERATE, generation_node)
builder.add_node(REFLECT, reflection_node)
builder.set_entry_point(GENERATE)


def should_continue(state: List[BaseMessage]):
    if len(state) > 6:
        return END
    return REFLECT


builder.add_conditional_edges("generate", should_continue, {END:END, "reflect":"reflect"})
builder.add_edge(REFLECT, GENERATE)
graph = builder.compile()
graph.get_graph().draw_mermaid_png(output_file_path="static/reflection_agent_graph.png")
graph.get_graph().print_ascii()
```

```python
request = HumanMessage(
    content="HÃ£y viáº¿t má»™t bÃ i Ä‘Äƒng trÃªn Twitter vá» lÃ½ do táº¡i sao HoÃ ng tá»­ bÃ© váº«n cÃ²n phÃ¹ há»£p vá»›i tuá»•i thÆ¡ hiá»‡n Ä‘áº¡i."
)
response = graph.invoke(request)
response
```
